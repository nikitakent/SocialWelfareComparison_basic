```python
import matplotlib.pyplot as plt
from scipy.integrate import quad
import numpy as np

# Defining the lambda parameter range
lambdas = np.linspace(0, 1, 500)

# SOCIAL WELFARE EQUATIONS

statquo_c00_eq = lambdas**2*((1**2/2) - ((lambdas/6+1/2)**2/2))
statquo_c005_eq = 2*lambdas*(1-lambdas)*((1**2/2 - 1/2) - (((lambdas/6+3/4)**2/2) - (lambdas/6+3/4)))

withPlat_c00_eq = lambdas**2*((1**2/2) - ((1/2)**2/2))
withPlat_c005_eq = 2*lambdas*(1-lambdas)*((1**2/2 - 1/2) - ((3/4)**2/2 - (3/4)))


statquo_socWef = statquo_c00_eq + statquo_c005_eq
withPlat_socWef = withPlat_c00_eq + withPlat_c005_eq

# difference_result = 0

# for i in lambdas:
#     difference = withPlat_socWef - statquo_socWef 
#     difference_result += difference

# print(difference_result)

# PLOT
plt.figure(figsize=(10, 6))
plt.plot(lambdas, withPlat_socWef, color='blue', label='with Platform')
plt.plot(lambdas, statquo_socWef, color='purple', label='Status quo')

# PLOT LABELS AND TITLES
plt.xlabel('Lambda (λ)')
plt.ylabel('Value')
plt.title('Social Welfare Comparison')
plt.legend()

# DISPLAY
plt.show()
```

```python
# Correcting the equations
revenue_newmech = (1/2)*(1/2)*lambdas**2 + (3/4-1/2)*(1/4)*(2*lambdas*(1-lambdas))
revenue_statusquo = (lambdas/6+1/2)*(1/2-lambdas/6)*lambdas**2 + ((lambdas/6)+1/4)*(1/4 -lambdas/6)*(2*lambdas*(1-lambdas))

# Creating the plot
plt.figure(figsize=(10, 6))
plt.plot(lambdas, revenue_newmech, color='blue', label='New mechanism')
plt.plot(lambdas, revenue_statusquo, color='purple', label='Status quo')


# Adding labels and title
plt.xlabel('Lambda (λ)')
plt.ylabel('Combined profit')
plt.title('Expected Profit Comparison')
plt.legend()

# Display the plot
plt.show()
```
